# News Headlines Sentiment

Use the news api to pull the latest news articles for bitcoin and ethereum and create a DataFrame of sentiment scores for each coin. 

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Initial imports
import os
import pandas as pd

from dotenv import load_dotenv

import nltk
nltk.download('vader_lexicon')

from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

%matplotlib inline
```

```python
# Read your api key environment variable
from newsapi import NewsApiClient as nac
```

```python
# Create a newsapi client
load_dotenv()
newsapi = nac(api_key=os.getenv('NEWS_API_KEY'))
print(newsapi)
```

```python
# Fetch the Bitcoin news articles
bitcoin = newsapi.get_everything(q='bitcoin', language='en', sort_by='relevancy')
```

```python
# Fetch the Ethereum news articles
ethereum = newsapi.get_everything(q='ethereum', language='en', sort_by='relevancy')
```

```python
# Create the Bitcoin sentiment scores DataFrame
bitcoin_senti = []

for article in bitcoin['articles']:
    try:
        text = article['content']
        senti = analyzer.polarity_scores(text)
        compound = senti['compound']
        pos = senti['pos']
        neu = senti['neu']
        neg = senti['neg']
        
        bitcoin_senti.append({
            'Text': text,
            'Compound': compound,
            'Positive': pos,
            'Neutral': neu,
            'Negative': neg
        })
    except AttributeError:
        pass

# DataFrame
bitcoin_df = pd.DataFrame(bitcoin_senti)
bitcoin_df.head()
```

```python
# Create the ethereum sentiment scores DataFrame
ethereum_senti = []

for article in ethereum['articles']:
    try:
        text = article['content']
        senti = analyzer.polarity_scores(text)
        compound = senti['compound']
        pos = senti['pos']
        neu = senti['neu']
        neg = senti['neg']
        
        ethereum_senti.append({
            'Text': text,
            'Compound': compound,
            'Positive': pos,
            'Neutral': neu,
            'Negative': neg
        })
    except AttributeError:
        pass

# DataFrame
ethereum_df = pd.DataFrame(ethereum_senti)
ethereum_df.head()
```

```python
# Describe the Bitcoin Sentiment
bitcoin_df.describe()
```

```python
# Describe the Ethereum Sentiment
ethereum_df.describe()
```

# Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word
2. Remove Punctuation
3. Remove Stopwords

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
# Expand the default stopwords list if necessary
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# Complete the tokenizer function
def tokenizer(text):
    """Tokenizes text."""
        
    # Remove punctuation
    regex = re.compile("[^a-zA-Z ]")
    re_clean = regex.sub('', text)
    words = word_tokenize(re_clean)
    
    # Initialize lemmatizer and stopwords
    lemmatizer = WordNetLemmatizer()
    sw = set(stopwords.words('english'))
    
    # Lemmatizing, removing stopwords, and tokenizing
    tokens = [lemmatizer.lemmatize(word.lower()) for word in words if word.lower() not in sw]
    
    return tokens
```

```python
# Create a new tokens column for bitcoin
bitcoin_df['Tokens'] = bitcoin_df['Text'].apply(tokenizer)
bitcoin_df.head()
```

```python
# Create a new tokens column for ethereum
ethereum_df['Tokens'] = ethereum_df['Text'].apply(tokenizer)
ethereum_df.head()
```

# NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams
```

```python
# Flatten the sublist in Tokens column
bitcoin_tokens = [item for sublist in bitcoin_df.Tokens.to_list() for item in sublist]

# Generate the Bitcoin N-grams where N=2
bitcoin_counts = Counter(ngrams(bitcoin_tokens, n=2))
bitcoin_counts.most_common(20)
```

```python
# Flatten the sublist in Tokens column
ethereum_tokens = [item for sublist in ethereum_df.Tokens.to_list() for item in sublist]

# Generate the Ethereum N-grams where N=2
ethereum_counts = Counter(ngrams(ethereum_tokens, n=2))
ethereum_counts.most_common(20)
```

```python
# Use the token_count function to generate the top 10 words from each coin
def token_count(tokens, N=10):
    """Returns the top N tokens from the frequency count"""
    return Counter(tokens).most_common(N)
```

```python
# Get the top 10 words for Bitcoin
token_count(bitcoin_tokens)
```

```python
# Get the top 10 words for Ethereum
token_count(ethereum_tokens)
```

# Named Entity Recognition

In this section, you will build a named entity recognition model for both coins and visualize the tags using SpaCy.

```python
import spacy
from spacy import displacy
```

```python
# Optional - download a language model for SpaCy
# !python -m spacy download en_core_web_sm
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')
```

## Bitcoin NER

```python
# Concatenate all of the bitcoin text together
bitcoin_join = ' '.join(bitcoin_df.Text.to_list())
print(bitcoin_join)
```

```python
# Run the NER processor on all of the text
bitcoin_ner = nlp(bitcoin_join)

# Add a title to the document
bitcoin_ner.user_data['title'] = 'Bitcoin NER'
```

```python
# Render the visualization
displacy.render(bitcoin_ner, style='dep')
```

```python
# List all Entities
for entity in bitcoin_ner.ents:
    print(entity.text, entity.label_)
```
